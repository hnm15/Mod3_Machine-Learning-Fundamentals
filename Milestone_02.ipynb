### Part 1: Baseline Modeling [3 pts]

Apply the following regression models to the scaled training dataset using **default parameters**:

- Linear Regression
- Ridge Regression
- Lasso Regression
- Decision Tree Regression
- Bagging

For each model:

- Report the **mean and standard deviation of CV RMSE Score** across all folds in a table. 


```python
# Add as many code cells as you need

#Linear Regression

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, RepeatedKFold


model = LinearRegression()

cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

linear_reg_results = {
    'Model': 'Linear Regression',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}


```

```python
# Ridge

from sklearn.linear_model import Ridge

model = Ridge()

cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

ridge_reg_results = {
    'Model': 'Ridge Regression',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}
```

```python
# Lasso
from sklearn.linear_model import Lasso

model = Lasso()

cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

lasso_reg_results = {
    'Model': 'Lasso Regression',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}
```

```python
# Decision Tree 

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

decision_tree_results = {
    'Model': 'Decision Tree Regressor',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}
```

```python
# Bagging 

from sklearn.ensemble import BaggingRegressor

model = BaggingRegressor()
cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

bagging_results = {
    'Model': 'Bagging Regressor',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}
```

```python
# Random Forest 

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=10, max_depth=10, n_jobs=-1, random_state=42)

cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

random_forest_results = {
    'Model': 'Random Forest Regressor',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}
```

```python
#Gradient Boosting

from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(random_state=42)

cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)

cv_results = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

mean_rmse = -np.mean(cv_results)
std_rmse = np.std(cv_results)

gradient_boosting_results = {
    'Model': 'Gradient Boosting Regressor',
    'Mean RMSE': mean_rmse,
    'Std RMSE': std_rmse
}
```

```python
results = [
    linear_reg_results,
    ridge_reg_results,
    lasso_reg_results,
    decision_tree_results,
    bagging_results,
    random_forest_results,
    gradient_boosting_results
]

results_df = pd.DataFrame(results)
print(results_df)
```
